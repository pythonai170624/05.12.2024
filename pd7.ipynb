```python
import numpy as np
import matplotlib.pyplot as plt

###################### # solution for sinq cos of matzeget 23 final exercise

fig = plt.figure(figsize=(7,2)) 

x = np.arange(0, 5, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

ax1 = fig.add_axes([0, 0, 0.5, 1])  # Main axes
ax1.plot(x, y_sin, label='sin(x)', color='blue')
ax1.set_title('Sine Function')
ax1.set_xlabel('X Values')
ax1.set_ylabel('sin(x)')
ax1.legend()
ax1.grid(True)  # Optional: Add gridlines

# Step 6: Plot the cos() function on the second subplot
ax2 = fig.add_axes([0.6, 0, 0.5, 1])  # Smaller, inset axes
ax2.plot(x, y_cos, label='cos(x)', color='orange')
ax2.set_title('Cosine Function')
ax2.set_xlabel('X Values')
ax2.set_ylabel('cos(x)')
ax2.legend()
ax2.grid(False) 

plt.show() 
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 5, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Create figure and axes for subplots
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))  # 2 rows, 1 column

# Sine plot
ax[0].plot(x, y_sin, label='sin(x)', color='blue')
ax[0].set_title('Sine Function')
ax[0].set_xlabel('X Values')
ax[0].set_ylabel('sin(x)')
ax[0].legend()
ax[0].grid(True)

# Cosine plot
ax[1].plot(x, y_cos, label='cos(x)', color='red')
ax[1].set_title('Cosine Function')
ax[1].set_xlabel('X Values')
ax[1].set_ylabel('cos(x)')
ax[1].legend()
ax[1].grid(True)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(0, 5, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)
y_exp = np.exp(x)  # Example additional function

# Create figure and axes for subplots
fig, ax = plt.subplots(2, 2, figsize=(10, 8))  # 2 rows, 2 columns

# 0,0 0,1
# 1,0 1,1

# Sine plot
ax[0][0].plot(x, y_sin, label='sin(x)', color='blue')
ax[0, 0].set_title('Sine Function')
ax[0, 0].set_xlabel('X Values')
ax[0, 0].set_ylabel('sin(x)')
ax[0, 0].legend()
ax[0, 0].grid(True)

# Cosine plot
ax[0, 1].plot(x, y_cos, label='cos(x)', color='red')
ax[0, 1].set_title('Cosine Function')
ax[0, 1].set_xlabel('X Values')
ax[0, 1].set_ylabel('cos(x)')
ax[0, 1].legend()
ax[0, 1].grid(True)

# Tangent plot
ax[1, 0].plot(x, y_tan, label='tan(x)', color='green')
ax[1, 0].set_title('Tangent Function')
ax[1, 0].set_xlabel('X Values')
ax[1, 0].set_ylabel('tan(x)')
ax[1, 0].legend()
ax[1, 0].grid(True)

# Exponential plot
# e = 2.718
ax[1, 1].plot(x, y_exp, label='exp(x)', color='purple')
ax[1, 1].set_title('Exponential Function')
ax[1, 1].set_xlabel('X Values')
ax[1, 1].set_ylabel('exp(x)')
ax[1, 1].legend()
ax[1, 1].grid(True)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create subplots
fig, axes = plt.subplots(nrows=4, ncols=4)

# Data
x = np.arange(10)
y = x * 2

# Get the shape of axes array
shape = axes.shape

# Loop through rows and columns of the axes
for i in range(shape[0]):  # Loop over rows
    for j in range(shape[1]):  # Loop over columns
        axes[i, j].plot(x, y)  # Plot data on each subplot
        axes[i, j].set_title(f'Plot[{i},{j}]')  # Set title for each subplot



# custom spaces
#fig.subplots_adjust(hspace=1.1, wspace=0.9) 

fig.suptitle('Big title', fontsize=14, y=1.0)

fig.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)

# Create the figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Plot with all the mentioned features
# First line: customized color, line width (lw), line style (ls), marker, and marker size
ax.plot(x, y, label='Sine Wave', color='#1FEBDD', lw=2, ls='--', marker='o', markersize=6)

# Second line: another set of customization
ax.plot(x, y2, label='Cosine Wave', color='red', lw=3, linestyle='-', marker='x', markersize=8)

# Set title, labels, and legend
ax.set_title('Customized Chart', fontsize=14)
ax.set_xlabel('X-axis', fontsize=12)
ax.set_ylabel('Y-axis', fontsize=12)

# Add legend
ax.legend()

# Add grid
ax.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Line Plot Data
x = np.arange(10)
y_linear = x * 2

# Scatter Plot Data
y_scatter = np.random.randn(10)

# Bar Plot Data
bar_categories = ['A', 'B', 'C', 'D', 'E']
bar_values = np.random.rand(5)

# Histogram Data
y_hist = np.random.randn(1000)

# Create Subplots
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

# Line Plot
ax[0, 0].plot(x, y_linear)
ax[0, 0].set_title('Line Plot')

# Scatter Plot
ax[0, 1].scatter(x, y_scatter, color='r')
ax[0, 1].set_title('Scatter Plot')

# Bar Plot
ax[1, 0].bar(bar_categories, bar_values, color='g')
ax[1, 0].set_title('Bar Plot')


# Histogram
ax[1, 1].hist(y_hist, bins=30, color='b', alpha=0.7)
ax[1, 1].set_title('Histogram')

# Set the title for the figure
fig.suptitle('4 Types of Matplotlib Plots', fontsize=16)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
